```python
# 각 행을 문장으로 변환  # 구분해주는게 더 도움됨
def row_to_sentence(row):
    return (
        f"[정책 제목: '{row['Policy Title']}'], "
        f"[정책 설명: '{row['Description']}'], "
        f"[거주지 및 소득: '{row['거주지 및 소득']}'], "
        f"[기타 유익 정보: '{row['기타 유익 정보']}'], "
        f"[비고: '{row['비고']}'], "
        f"[사업 신청 기간: '{row['사업 신청 기간']}'], "
        f"[사업 운영 기간: '{row['사업 운영 기간']}'], "
        f"[신청 절차: '{row['신청 절차']}'], "
        f"[연령: '{row['연령']}'],"
        f"[운영 기관: '{row['운영 기관']}'],"
        f"[주관 기관: '{row['주관 기관']}'],"
        f"[지원 규모: '{row['지원 규모(명)']}'],"
        f"[지원 내용: '{row['지원 내용']}'],"
        f"[참여 제한 대상: '{row['참여 제한 대상']}'],"
        f"[취업 상태: '{row['취업 상태']}'],"
        f"[특화 분야: '{row['특화 분야']}'],"
        f"[학력: '{row['학력']}']"
    )


# 결과 생성
sentences = data5.apply(row_to_sentence, axis=1)

# 결과 출력
for sentence in sentences:
    print(sentence)
```

```python
# sentences를 DataFrame으로 변환
sentences_df = pd.DataFrame(sentences, columns=["sentence"])

sentences_df["index"] = range(1, len(sentences_df) + 1)  # 1부터 시작하는 순번 생성

# CSV 파일로 저장
output_file = "policy_saving_sentences.csv"
sentences_df.to_csv(output_file, index=False, encoding="utf-8-sig")

print(f"'{output_file}' 파일로 저장되었습니다.")
```

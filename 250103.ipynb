```python
import json
from dataclasses import dataclass
from typing import List, Optional, Dict, Any
import re
from datetime import datetime, date

debug=True

@dataclass
class MonetaryValue:
    amount: int  # 원 단위
    original_text: str

@dataclass
class Period:
    start_date: Optional[date]
    end_date: Optional[date]
    original_text: str

@dataclass
class IncomeRequirement:
    type: str  # "중위소득", "도시근로자" 등
    percentage: Optional[float]  # 비율 (예: 120.0)
    amount: Optional[MonetaryValue]  # 구체적 금액이 명시된 경우
    household_type: str  # "원가구", "청년가구", "부부합산" 등

@dataclass
class PolicyCriteria:
    age_min: Optional[int]
    age_max: Optional[int]
    region: List[str]
    marriage_required: Optional[bool]
    residence_status: Dict[str, Any]
    income_requirements: List[IncomeRequirement]
    asset_limit: Optional[MonetaryValue]
    application_period: Optional[Period]
    support_amount: Optional[MonetaryValue]
    requirements: Dict[str, bool]
    validation_errors: List[str]
    income_assets: Dict[str, Any]
    status: Dict[str, bool]

    def __str__(self):
        return (
            f"PolicyCriteria(\n"
            f"  age_min: {self.age_min},\n"
            f"  age_max: {self.age_max},\n"
            f"  region: {', '.join(self.region) if self.region else 'None'},\n"
            f"  marriage_required: {self.marriage_required},\n"
            f"  residence_status: {self.residence_status},\n"
            f"  income_requirements: {self.income_requirements},\n"
            f"  asset_limit: {self.asset_limit},\n"
            f"  application_period: {self.application_period},\n"
            f"  support_amount: {self.support_amount},\n"
            f"  requirements: {self.requirements},\n"
            f"  validation_errors: {self.validation_errors},\n"
            f"  income_assets: {self.income_assets},\n"
            f"  status: {self.status}\n"
            f")"
        )

class KoreanNumberConverter:
    def __init__(self):
        self.korean_numbers = {
            '일': 1, '이': 2, '삼': 3, '사': 4, '오': 5,
            '육': 6, '칠': 7, '팔': 8, '구': 9,
            '일': 1, '둘': 2, '셋': 3, '넷': 4, '다섯': 5,
            '여섯': 6, '일곱': 7, '여덟': 8, '아홉': 9,
            '십': 10, '백': 100, '천': 1000, '만': 10000,
            '억': 100000000
        }
        
    def convert_to_number(self, korean_text: str) -> Optional[int]:
        try:
            # 숫자로 된 경우 바로 변환
            return int(korean_text)
        except ValueError:
            result = 0
            temp = 0
            unit = 1
            
            for char in korean_text:
                if char in self.korean_numbers:
                    number = self.korean_numbers[char]
                    if number >= 10000:
                        temp = temp if temp > 0 else 1
                        result += (temp * number)
                        temp = 0
                    elif number >= 10:
                        temp = temp if temp > 0 else 1
                        temp *= number
                    else:
                        temp = number
            
            return result + temp

class MonetaryParser:
    def __init__(self):
        self.number_converter = KoreanNumberConverter()
        self.amount_patterns = [
            r'(\d+|\w+)(?:백만|천만|만|억)원',
            r'(\d+|\w+)원',
            r'금?\s*(\d+|\w+)(?:백만|천만|만|억)\s*원?'
        ]
        
    def parse_amount(self, text: str) -> Optional[MonetaryValue]:
        for pattern in self.amount_patterns:
            match = re.search(pattern, text)
            if match:
                amount_text = match.group(1)
                unit_text = match.group(0)
                
                base_amount = self.number_converter.convert_to_number(amount_text)
                if not base_amount:
                    continue
                    
                # 단위 처리
                if '억' in unit_text:
                    amount = base_amount * 100000000
                elif '천만' in unit_text:
                    amount = base_amount * 10000000
                elif '백만' in unit_text:
                    amount = base_amount * 1000000
                elif '만' in unit_text:
                    amount = base_amount * 10000
                else:
                    amount = base_amount
                    
                return MonetaryValue(amount=amount, original_text=match.group(0))
        
        return None

class DateParser:
    def __init__(self):
        self.date_patterns = [
            r'(\d{4})[./\s-](\d{1,2})[./\s-](\d{1,2})',  # 2024-01-01
            r'(\d{4})년\s*(\d{1,2})월\s*(\d{1,2})일',    # 2024년 1월 1일
            r'(\d{2})\.(\d{1,2})\.(\d{1,2})',           # 24.1.1
        ]
        self.period_patterns = [
            r'(?:신청|접수|모집)기간\s*:?\s*([^~]*)[~까지]([^\n]*)',
            r'(\d{4}[./년\s]\d{1,2}[./월\s]\d{1,2}[일]?\s*)[~부터\s]+(\d{4}[./년\s]\d{1,2}[./월\s]\d{1,2}[일]?)',
        ]
        
    def parse_date(self, text: str) -> Optional[date]:
        for pattern in self.date_patterns:
            match = re.search(pattern, text)
            if match:
                year = int(match.group(1))
                month = int(match.group(2))
                day = int(match.group(3))
                
                # 2자리 연도 처리
                if year < 100:
                    year += 2000
                    
                try:
                    return date(year, month, day)
                except ValueError:
                    continue
        
        return None
        
    def parse_period(self, text: str) -> Optional[Period]:
        for pattern in self.period_patterns:
            match = re.search(pattern, text)
            if match:
                start_text = match.group(1)
                end_text = match.group(2)
                
                start_date = self.parse_date(start_text)
                end_date = self.parse_date(end_text)
                
                return Period(
                    start_date=start_date,
                    end_date=end_date,
                    original_text=match.group(0)
                )
        
        return None

class IncomeParser:
    def __init__(self):
        self.patterns = {
            "중위소득": r'중위소득\s*(\d+)%\s*(?:이하|미만)?',
            "도시근로자": r'도시근로자\s*(?:월평균\s*)?소득\s*(\d+)%\s*(?:이하|미만)?'
        }
        
    def parse_income_requirements(self, text: str) -> List[IncomeRequirement]:
        requirements = []
        
        for income_type, pattern in self.patterns.items():
            matches = re.finditer(pattern, text)
            for match in matches:
                percentage = float(match.group(1))
                
                # 가구 유형 확인
                household_type = "일반"
                if "원가구" in text:
                    household_type = "원가구"
                elif "부부합산" in text:
                    household_type = "부부합산"
                elif "청년" in text:
                    household_type = "청년가구"
                
                requirements.append(IncomeRequirement(
                    type=income_type,
                    percentage=percentage,
                    amount=None,  # 구체적 금액은 별도 처리 필요
                    household_type=household_type
                ))
                
        return requirements



class RequirementsValidator:
    def validate_requirements(self, req: PolicyCriteria) -> List[str]:
        errors = []
        
        # 나이 검증
        if req.age_min is not None and req.age_max is not None:
            if req.age_min > req.age_max:
                errors.append(f"최소 나이({req.age_min})가 최대 나이({req.age_max})보다 큽니다")
                
        # 날짜 검증
        if req.application_period:
            if (req.application_period.start_date and 
                req.application_period.end_date and 
                req.application_period.start_date > req.application_period.end_date):
                errors.append("시작일이 종료일보다 늦습니다")
                
        # 필수 항목 검증
        if not req.region and not req.requirements.get("전국대상", False):
            errors.append("지역 정보가 없습니다")
            
        return errors


class RegionManager:
    def __init__(self):
        self.region_dict = {
            
"경상북도": [
            "포항시 남구", "포항시 북구", "경주시", "김천시", "안동시", "구미시", "영주시", 
            "영천시", "상주시", "문경시", "경산시", "군위군", "의성군", "청송군", "영양군",
            "영덕군", "청도군", "고령군", "성주군", "칠곡군", "예천군", "봉화군", "울진군",
            "울릉군"
        ],
        "대전광역시": ["동구", "중구", "서구", "유성구", "대덕구"],
        "서울특별시": [
            "종로구", "중구", "용산구", "성동구", "광진구", "동대문구", "중랑구", "성북구",
            "강북구", "도봉구", "노원구", "은평구", "서대문구", "마포구", "양천구", "강서구",
            "구로구", "금천구", "영등포구", "동작구", "관악구", "서초구", "강남구", "송파구",
            "강동구"
        ],
        "부산광역시": [
            "중구", "서구", "동구", "영도구", "부산진구", "동래구", "남구", "북구",
            "해운대구", "사하구", "금정구", "강서구", "연제구", "수영구", "사상구",
            "기장군"
        ],
        "인천광역시": [
            "중구", "동구", "미추홀구", "연수구", "남동구", "부평구", "계양구", "서구",
            "강화군", "옹진군"
        ],
        "광주광역시": ["동구", "서구", "남구", "북구", "광산구"],
        "대구광역시": [
            "중구", "동구", "서구", "남구", "북구", "수성구", "달서구", "달성군"
        ],
        "울산광역시": ["중구", "남구", "동구", "북구", "울주군"],
        "세종특별자치시": ["세종시"],
        "제주특별자치도": ["제주시", "서귀포시"],
        "기타": ["기타"]
        }
        
        # Create reverse lookup dictionary
        self.district_to_region = {}
        for region, districts in self.region_dict.items():
            for district in districts:
                self.district_to_region[district] = region
                # Also add the district name without the 시/군/구
                district_name = re.sub(r'시|군|구$', '', district)
                self.district_to_region[district_name] = region

    def normalize_region(self, region_text: str) -> List[str]:
        """주어진 텍스트에서 지역 정보를 정규화하여 반환"""
        if not region_text:
            return []
            
        found_regions = []
        
        # 광역시/도 직접 매칭
        for region in self.region_dict.keys():
            if region in region_text:
                found_regions.append(region)
                
        # 구/군 매칭
        for district in self.district_to_region.keys():
            if district in region_text:
                parent_region = self.district_to_region[district]
                if parent_region not in found_regions:
                    found_regions.append(parent_region)
                found_regions.append(district)
                
        return found_regions


class ExtendedRequirementsParser():
    def __init__(self):
        self.age_patterns = [
            r'(?:만\s*)?(\d+)\s*[~∼-]\s*(?:만\s*)?(\d+)\s*세?',  # 19-39세, 만 19~39세
            r'(?:만\s*)?(\d+)\s*세?\s*이[상하]',  # 19세 이상, 만 19세 이상
            r'청년\(\s*(?:만\s*)?(\d+)\s*[~∼-]\s*(?:만\s*)?(\d+)\s*세?\s*\)'  # 청년(19-39세)
        ]
        
        self.region_patterns = [
            r'([가-힣]+[시군구도])[에서\s]',  # ~시/군/구/도에서
            r'([가-힣]+[시군구도])[ 거주]',  # ~시/군/구/도 거주
            r'([가-힣]+[시군구도])\s',  # ~시/군/구/도
            r'지역기준[:\s]*([가-힣,\s]+)', 
        ]
        
        self.marriage_patterns = [
            r'신혼부부',
            r'혼인',
            r'결혼',
            r'미혼'
        ]
        self.region_manager = RegionManager()  # 이전 코드의 RegionManager 사용
        self.monetary_parser = MonetaryParser()
        self.date_parser = DateParser()
        self.income_parser = IncomeParser()
        self.validator = RequirementsValidator()
    def _get_detail(self, item: Dict, title: str) -> str:
        for detail in item["Details"]:
            if detail["Title"] == title:
                return detail["Content"]
        return ""


    def extract_age(self, text: str) -> tuple[Optional[int], Optional[int]]:
        for pattern in self.age_patterns:
            match = re.search(pattern, text)
            if match:
                if len(match.groups()) == 2:
                    return int(match.group(1)), int(match.group(2))
                else:
                    age = int(match.group(1))
                    if '이상' in text:
                        return age, None
                    elif '이하' in text:
                        return None, age
                    elif '대' in text:
                        return age, age + 9
        return None, None

    def extract_regions(self, text: str) -> List[str]:
        found_regions = set()
        
        # 패턴 매칭으로 지역 추출
        for pattern in self.region_patterns:
            matches = re.finditer(pattern, text)
            for match in matches:
                region_text = match.group(1).strip()
                # 쉼표로 구분된 여러 지역 처리
                for region in region_text.split(','):
                    region = region.strip()
                    normalized_regions = self.region_manager.normalize_region(region)
                    found_regions.update(normalized_regions)
        
        return list(found_regions)

    def extract_marriage_status(self, text: str) -> Optional[bool]:
        if '미혼' in text:
            return False
        elif any(keyword in text for keyword in ['신혼부부', '혼인', '결혼']):
            return True
        return None

    def extract_residence_status(self, text: str) -> Dict[str, Any]:
        status = {
            "무주택": False,
            "전월세": False,
            "임차보증금_제한": None,
            "월세_제한": None
        }
        
        if '무주택' in text:
            status["무주택"] = True
        
        if '전월세' in text or '월세' in text:
            status["전월세"] = True
            
        # 임차보증금 제한 추출
        match = re.search(r'임차보증금\s*(\d+)[만천억]*원', text)
        if match:
            amount = self._convert_korean_amount(match.group(1))
            status["임차보증금_제한"] = amount
            
        # 월세 제한 추출
        match = re.search(r'월세\s*(\d+)[만천]*원', text)
        if match:
            amount = self._convert_korean_amount(match.group(1))
            status["월세_제한"] = amount
            
        return status

    def extract_income_assets(self, text: str) -> Dict[str, Any]:
        income_assets = {
            "소득_기준": None,
            "재산_기준": None,
            "기타_소득조건": []
        }
        
        # 중위소득 기준 추출
        matches = re.finditer(r'중위소득\s*(\d+)%', text)
        if matches:
            income_assets["소득_기준"] = [f"중위소득 {m.group(1)}%" for m in matches]
            
        # 재산 기준 추출
        match = re.search(r'재산\s*(\d+)(?:백만|천만|억)원', text)
        if match:
            income_assets["재산_기준"] = self._convert_korean_amount(match.group(1))
            
        return income_assets

    def _convert_korean_amount(self, amount_str: str) -> int:
        # 숫자 변환 로직 구현
        try:
            return int(amount_str)
        except ValueError:
            return 0

    def extract_status(self, text: str) -> Dict[str, bool]:
        status = {
            "대학생": False,
            "직장인": False,
            "창업자": False,
            "귀농인": False
        }
        
        if '대학생' in text or '재학생' in text:
            status["대학생"] = True
        if '직장' in text or '근로자' in text:
            status["직장인"] = True
        if '창업' in text:
            status["창업자"] = True
        if '귀농' in text:
            status["귀농인"] = True
            
        return status

    def extract_requirements(self, text: str) -> Dict[str, bool]:
        requirements = {
            "청약통장": False,
            "전입신고": False,
            "부모별도거주": False
        }
        
        if '청약통장' in text:
            requirements["청약통장"] = True
        if '전입신고' in text:
            requirements["전입신고"] = True
        if '부모' in text and ('별도' in text or '따로' in text):
            requirements["부모별도거주"] = True
            
        return requirements

    
    def parse(self, item: Dict) -> PolicyCriteria:
        # 기본 정보 추출 (이전 코드의 기능 활용)
        text = self._get_detail(item, "거주지 및 소득")
        age_min, age_max = self.extract_age(text)
        regions = self.extract_regions(text)
        marriage_required = self.extract_marriage_status(text)
        residence_status = self.extract_residence_status(text)
        income_assets = self.extract_income_assets(text)
        status = self.extract_status(text)
        requirementss = self.extract_requirements(text)
        
        # 기타 조건은 향후 확장성을 위해 리스트로 유지
        # other = []
        
        # 새로운 정보 추출
        income_requirements = self.income_parser.parse_income_requirements(text)
        application_period = self.date_parser.parse_period(text)
        
        # 지원금액 추출
        support_amount = None
        if "지원금액" in text or "지원한도" in text:
            amount_text = text[text.find("지원금액"):]
            support_amount = self.monetary_parser.parse_amount(amount_text)
        
        # 요구사항 생성
        requirements = PolicyCriteria(
            age_min=age_min,
            age_max=age_max,
            region=regions,
            marriage_required=marriage_required,
            residence_status=residence_status,  # 이전 코드의 기능 활용
            income_requirements=income_requirements,
            asset_limit=None,  # 자산 제한 추출 필요
            application_period=application_period,
            support_amount=support_amount,
            requirements=requirementss,  # 이전 코드의 기능 활용
            validation_errors=[],
            income_assets=income_assets,
            status=status,
        )
        
        if debug:
            print(text)
            print(requirements)
            print('-'*50)

        # 검증 수행
        requirements.validation_errors = self.validator.validate_requirements(requirements)
        
        return requirements


@dataclass
class Policy:
    policy_id: str
    title: str
    criteria: PolicyCriteria
    description: str
    details: Dict[str, str]

class PolicyFilter:
    def __init__(self, json_data: str):
        self.policies = self._parse_policies(json_data)
    
    def _parse_policies(self, json_data: str) -> List[Policy]:
        data = json.loads(json_data)
        policies = []
        parser = ExtendedRequirementsParser()
        
        for item in data:
            criteria=parser.parse(item)
            
            policy = Policy(
                policy_id=self._get_detail(item, "정책 번호"),
                title=item["Policy Title"],
                criteria=criteria,
                description=item["Description"],
                details={detail["Title"]: detail["Content"] for detail in item["Details"]}
            )
            policies.append(policy)
            
        
        return policies
    
    def _get_detail(self, item: Dict, title: str) -> str:
        for detail in item["Details"]:
            if detail["Title"] == title:
                return detail["Content"]
        return ""
    
    
    def filter_policies(self, 
                       age: int = None,
                       region: str = None,
                       income_percent: float = None,
                       monthly_rent: int = None,
                       deposit: int = None,
                       is_married: bool = None) -> List[Policy]:
        """
        주어진 조건에 맞는 정책들을 필터링하여 반환합니다.
        
        Args:
            age: 나이
            region: 지역
            income_percent: 중위소득 대비 비율 (%)
            monthly_rent: 월세 금액 (만원)
            deposit: 보증금 금액 (만원)
            is_married: 혼인 여부
        """
        eligible_policies = []
        
        for policy in self.policies:
            criteria = policy.criteria
            
            # 각 조건 검사
            if age is not None and criteria.age_min is not None and criteria.age_max is not None:
                if not (criteria.age_min <= age <= criteria.age_max):
                    continue
                    
            # if region is not None and criteria.region is not None:
            #     if criteria.region != region:
            #         continue
                    
            # if income_percent is not None and criteria.income is not None:
            #     if income_percent > criteria.income_limit:
            #         continue
                    
            # if monthly_rent is not None and criteria.monthly_rent_limit is not None:
            #     if monthly_rent > criteria.monthly_rent_limit:
            #         continue
                    
            # if deposit is not None and criteria.deposit_limit is not None:
            #     if deposit > criteria.deposit_limit:
            #         continue
                    
            # if is_married is not None and criteria.marriage_required is not None:
            #     if criteria.marriage_required and not is_married:
            #         continue
            
            eligible_policies.append(policy)
            
        return eligible_policies

def main():
    # 사용 예시
    with open('filtered_policies.json', 'r', encoding='utf-8') as f:
        json_data = f.read()
    
    policy_filter = PolicyFilter(json_data)
    
    # 예시: 28세, 경북 거주, 중위소득 100%, 월세 50만원, 보증금 3000만원, 기혼
    eligible_policies = policy_filter.filter_policies(
        age=28,
        region="경북",
        income_percent=100,
        monthly_rent=50,
        deposit=3000,
        is_married=True
    )

    print(len(eligible_policies))
    
    for policy in eligible_policies:
        print(f"\n정책명: {policy.title}")
        print(f"정책 ID: {policy.policy_id}")
        print(f"지원 내용: {policy.details.get('거주지 및 소득', '')}")

if __name__ == "__main__":
    main()
```

```python
import json
from dataclasses import dataclass
from typing import List, Optional, Dict, Any
import re
from datetime import datetime, date

debug=True

@dataclass
class MonetaryValue:
    amount: int  # 원 단위
    original_text: str

@dataclass
class Period:
    start_date: Optional[date]
    end_date: Optional[date]
    original_text: str

@dataclass
class IncomeRequirement:
    type: str  # "중위소득", "도시근로자" 등
    percentage: Optional[float]  # 비율 (예: 120.0)
    amount: Optional[MonetaryValue]  # 구체적 금액이 명시된 경우
    household_type: str  # "원가구", "청년가구", "부부합산" 등

@dataclass
class PolicyCriteria:
    age_min: Optional[int]
    age_max: Optional[int]
    region: List[str]
    marriage_required: Optional[bool]
    residence_status: Dict[str, Any]
    income_requirements: List[IncomeRequirement]
    asset_limit: Optional[MonetaryValue]
    application_period: Optional[Period]
    support_amount: Optional[MonetaryValue]
    requirements: Dict[str, bool]
    validation_errors: List[str]
    income_assets: Dict[str, Any]
    status: Dict[str, bool]

    def __str__(self):
        return (
            f"PolicyCriteria(\n"
            f"  age_min: {self.age_min},\n"
            f"  age_max: {self.age_max},\n"
            f"  region: {', '.join(self.region) if self.region else 'None'},\n"
            f"  marriage_required: {self.marriage_required},\n"
            f"  residence_status: {self.residence_status},\n"
            f"  income_requirements: {self.income_requirements},\n"
            f"  asset_limit: {self.asset_limit},\n"
            f"  application_period: {self.application_period},\n"
            f"  support_amount: {self.support_amount},\n"
            f"  requirements: {self.requirements},\n"
            f"  validation_errors: {self.validation_errors},\n"
            f"  income_assets: {self.income_assets},\n"
            f"  status: {self.status}\n"
            f")"
        )

class KoreanNumberConverter:
    def __init__(self):
        self.korean_numbers = {
            '일': 1, '이': 2, '삼': 3, '사': 4, '오': 5,
            '육': 6, '칠': 7, '팔': 8, '구': 9,
            '일': 1, '둘': 2, '셋': 3, '넷': 4, '다섯': 5,
            '여섯': 6, '일곱': 7, '여덟': 8, '아홉': 9,
            '십': 10, '백': 100, '천': 1000, '만': 10000,
            '억': 100000000
        }
        
    def convert_to_number(self, korean_text: str) -> Optional[int]:
        try:
            # 숫자로 된 경우 바로 변환
            return int(korean_text)
        except ValueError:
            result = 0
            temp = 0
            unit = 1
            
            for char in korean_text:
                if char in self.korean_numbers:
                    number = self.korean_numbers[char]
                    if number >= 10000:
                        temp = temp if temp > 0 else 1
                        result += (temp * number)
                        temp = 0
                    elif number >= 10:
                        temp = temp if temp > 0 else 1
                        temp *= number
                    else:
                        temp = number
            
            return result + temp

class MonetaryParser:
    def __init__(self):
        self.number_converter = KoreanNumberConverter()
        self.amount_patterns = [
            r'(\d+|\w+)(?:백만|천만|만|억)원',
            r'(\d+|\w+)원',
            r'금?\s*(\d+|\w+)(?:백만|천만|만|억)\s*원?'
        ]
        
    def parse_amount(self, text: str) -> Optional[MonetaryValue]:
        for pattern in self.amount_patterns:
            match = re.search(pattern, text)
            if match:
                amount_text = match.group(1)
                unit_text = match.group(0)
                
                base_amount = self.number_converter.convert_to_number(amount_text)
                if not base_amount:
                    continue
                    
                # 단위 처리
                if '억' in unit_text:
                    amount = base_amount * 100000000
                elif '천만' in unit_text:
                    amount = base_amount * 10000000
                elif '백만' in unit_text:
                    amount = base_amount * 1000000
                elif '만' in unit_text:
                    amount = base_amount * 10000
                else:
                    amount = base_amount
                    
                return MonetaryValue(amount=amount, original_text=match.group(0))
        
        return None

class DateParser:
    def __init__(self):
        self.date_patterns = [
            r'(\d{4})[./\s-](\d{1,2})[./\s-](\d{1,2})',  # 2024-01-01
            r'(\d{4})년\s*(\d{1,2})월\s*(\d{1,2})일',    # 2024년 1월 1일
            r'(\d{2})\.(\d{1,2})\.(\d{1,2})',           # 24.1.1
        ]
        self.period_patterns = [
            r'(?:신청|접수|모집)기간\s*:?\s*([^~]*)[~까지]([^\n]*)',
            r'(\d{4}[./년\s]\d{1,2}[./월\s]\d{1,2}[일]?\s*)[~부터\s]+(\d{4}[./년\s]\d{1,2}[./월\s]\d{1,2}[일]?)',
        ]
        
    def parse_date(self, text: str) -> Optional[date]:
        for pattern in self.date_patterns:
            match = re.search(pattern, text)
            if match:
                year = int(match.group(1))
                month = int(match.group(2))
                day = int(match.group(3))
                
                # 2자리 연도 처리
                if year < 100:
                    year += 2000
                    
                try:
                    return date(year, month, day)
                except ValueError:
                    continue
        
        return None
        
    def parse_period(self, text: str) -> Optional[Period]:
        for pattern in self.period_patterns:
            match = re.search(pattern, text)
            if match:
                start_text = match.group(1)
                end_text = match.group(2)
                
                start_date = self.parse_date(start_text)
                end_date = self.parse_date(end_text)
                
                return Period(
                    start_date=start_date,
                    end_date=end_date,
                    original_text=match.group(0)
                )
        
        return None

class IncomeParser:
    def __init__(self):
        self.patterns = {
            "중위소득": r'중위소득\s*(\d+)%\s*(?:이하|미만)?',
            "도시근로자": r'도시근로자\s*(?:월평균\s*)?소득\s*(\d+)%\s*(?:이하|미만)?'
        }
        
    def parse_income_requirements(self, text: str) -> List[IncomeRequirement]:
        requirements = []
        
        for income_type, pattern in self.patterns.items():
            matches = re.finditer(pattern, text)
            for match in matches:
                percentage = float(match.group(1))
                
                # 가구 유형 확인
                household_type = "일반"
                if "원가구" in text:
                    household_type = "원가구"
                elif "부부합산" in text:
                    household_type = "부부합산"
                elif "청년" in text:
                    household_type = "청년가구"
                
                requirements.append(IncomeRequirement(
                    type=income_type,
                    percentage=percentage,
                    amount=None,  # 구체적 금액은 별도 처리 필요
                    household_type=household_type
                ))
                
        return requirements



class RequirementsValidator:
    def validate_requirements(self, req: PolicyCriteria) -> List[str]:
        errors = []
        
        # 나이 검증
        if req.age_min is not None and req.age_max is not None:
            if req.age_min > req.age_max:
                errors.append(f"최소 나이({req.age_min})가 최대 나이({req.age_max})보다 큽니다")
                
        # 날짜 검증
        if req.application_period:
            if (req.application_period.start_date and 
                req.application_period.end_date and 
                req.application_period.start_date > req.application_period.end_date):
                errors.append("시작일이 종료일보다 늦습니다")
                
        # 필수 항목 검증
        if not req.region and not req.requirements.get("전국대상", False):
            errors.append("지역 정보가 없습니다")
            
        return errors


class RegionManager:
    def __init__(self):
        self.region_dict = {
            
"경상북도": [
            "포항시 남구", "포항시 북구", "경주시", "김천시", "안동시", "구미시", "영주시", 
            "영천시", "상주시", "문경시", "경산시", "군위군", "의성군", "청송군", "영양군",
            "영덕군", "청도군", "고령군", "성주군", "칠곡군", "예천군", "봉화군", "울진군",
            "울릉군"
        ],
        "대전광역시": ["동구", "중구", "서구", "유성구", "대덕구"],
        "서울특별시": [
            "종로구", "중구", "용산구", "성동구", "광진구", "동대문구", "중랑구", "성북구",
            "강북구", "도봉구", "노원구", "은평구", "서대문구", "마포구", "양천구", "강서구",
            "구로구", "금천구", "영등포구", "동작구", "관악구", "서초구", "강남구", "송파구",
            "강동구"
        ],
        "부산광역시": [
            "중구", "서구", "동구", "영도구", "부산진구", "동래구", "남구", "북구",
            "해운대구", "사하구", "금정구", "강서구", "연제구", "수영구", "사상구",
            "기장군"
        ],
        "인천광역시": [
            "중구", "동구", "미추홀구", "연수구", "남동구", "부평구", "계양구", "서구",
            "강화군", "옹진군"
        ],
        "광주광역시": ["동구", "서구", "남구", "북구", "광산구"],
        "대구광역시": [
            "중구", "동구", "서구", "남구", "북구", "수성구", "달서구", "달성군"
        ],
        "울산광역시": ["중구", "남구", "동구", "북구", "울주군"],
        "세종특별자치시": ["세종시"],
        "제주특별자치도": ["제주시", "서귀포시"],
        "기타": ["기타"]
        }
        
        # Create reverse lookup dictionary
        self.district_to_region = {}
        for region, districts in self.region_dict.items():
            for district in districts:
                self.district_to_region[district] = region
                # Also add the district name without the 시/군/구
                district_name = re.sub(r'시|군|구$', '', district)
                self.district_to_region[district_name] = region

    def normalize_region(self, region_text: str) -> List[str]:
        """주어진 텍스트에서 지역 정보를 정규화하여 반환"""
        if not region_text:
            return []
            
        found_regions = []
        
        # 광역시/도 직접 매칭
        for region in self.region_dict.keys():
            if region in region_text:
                found_regions.append(region)
                
        # 구/군 매칭
        for district in self.district_to_region.keys():
            if district in region_text:
                parent_region = self.district_to_region[district]
                if parent_region not in found_regions:
                    found_regions.append(parent_region)
                found_regions.append(district)
                
        return found_regions


class ExtendedRequirementsParser():
    def __init__(self):
        self.age_patterns = [
            r'(?:만\s*)?(\d+)\s*[~∼-]\s*(?:만\s*)?(\d+)\s*세?',  # 19-39세, 만 19~39세
            r'(?:만\s*)?(\d+)\s*세?\s*이[상하]',  # 19세 이상, 만 19세 이상
            r'청년\(\s*(?:만\s*)?(\d+)\s*[~∼-]\s*(?:만\s*)?(\d+)\s*세?\s*\)',  # 청년(19-39세)
            r'(?:만\s*)?(\d+)\s*세\s*(?:이상|~)\s*(?:만\s*)?(\d+)\s*세\s*(?:이하|미만)',  # 18세 이상 39세 이하
            r'(?:만\s*)?(\d+)\s*세\s*~\s*(?:만\s*)?(\d+)\s*세',  # 18세 ~ 34세
        ]
        
        self.region_patterns = [
            r'([가-힣]+[시군구도])[에서\s]',  # ~시/군/구/도에서
            r'([가-힣]+[시군구도])[ 거주]',  # ~시/군/구/도 거주
            r'([가-힣]+[시군구도])\s',  # ~시/군/구/도
            r'지역기준[:\s]*([가-힣,\s]+)',
            r'([가-힣]+[시군구도])\s*내',  # 경북도 내
            r'([가-힣]+[시군구도])\s*소재',  # 대전시 소재
        ]
        
        self.income_patterns = [
            r'전년도\s*도시근로자\s*월평균소득의\s*(\d+)%',  # 전년도 도시근로자 월평균소득의 120%
            r'연소득\s*(?:\(부부합산\))?\s*(\d+)(?:천만|백만|억)원\s*이하',  # 연소득(부부합산) 6천만원 이하
            r'소득\s*:\s*(?:\(([^)]+)\))?\s*기준\s*중위소득\s*(\d+)%\s*이하',  # (청년독립가구) 기준 중위소득 60%이하
        ]
        
        # 기존 코드 유지
        self.marriage_patterns = [
            r'신혼부부',
            r'혼인',
            r'결혼',
            r'미혼'
        ]
        self.region_manager = RegionManager()
        self.monetary_parser = MonetaryParser()
        self.date_parser = DateParser()
        self.income_parser = IncomeParser()
        self.validator = RequirementsValidator()

    def _get_detail(self, item: Dict, title: str) -> str:
        for detail in item["Details"]:
            if detail["Title"] == title:
                return detail["Content"]
        return ""


    def extract_age(self, text: str) -> tuple[Optional[int], Optional[int]]:
        for pattern in self.age_patterns:
            match = re.search(pattern, text)
            if match:
                if len(match.groups()) == 2:
                    return int(match.group(1)), int(match.group(2))
                else:
                    age = int(match.group(1))
                    if '이상' in text:
                        return age, None
                    elif '이하' in text or '미만' in text:
                        return None, age
                    elif '대' in text:
                        return age, age + 9
        return None, None


    def extract_regions(self, text: str) -> List[str]:
        found_regions = set()
        
        for pattern in self.region_patterns:
            matches = re.finditer(pattern, text)
            for match in matches:
                region_text = match.group(1).strip()
                # 쉼표로 구분된 여러 지역 처리
                for region in region_text.split(','):
                    region = region.strip()
                    # 지역명이 2글자 이상인 경우에만 처리
                    if len(region) >= 2:
                        normalized_regions = self.region_manager.normalize_region(region)
                        found_regions.update(normalized_regions)
        
        return list(found_regions)

    def extract_marriage_status(self, text: str) -> Optional[bool]:
        if '미혼' in text:
            return False
        elif any(keyword in text for keyword in ['신혼부부', '혼인', '결혼']):
            return True
        return None

    def extract_residence_status(self, text: str) -> Dict[str, Any]:
        status = {
            "무주택": False,
            "전월세": False,
            "임차보증금_제한": None,
            "월세_제한": None
        }
        
        if '무주택' in text:
            status["무주택"] = True
        
        if '전월세' in text or '월세' in text:
            status["전월세"] = True
            
        # 임차보증금 제한 추출
        match = re.search(r'임차보증금\s*(\d+)[만천억]*원', text)
        if match:
            amount = self._convert_korean_amount(match.group(1))
            status["임차보증금_제한"] = amount
            
        # 월세 제한 추출
        match = re.search(r'월세\s*(\d+)[만천]*원', text)
        if match:
            amount = self._convert_korean_amount(match.group(1))
            status["월세_제한"] = amount
            
        return status

    def extract_income_assets(self, text: str) -> Dict[str, Any]:
        income_assets = {
            "소득_기준": [],
            "재산_기준": None,
            "기타_소득조건": []
        }
        
        # 중위소득 기준 추출
        for pattern in self.income_patterns:
            matches = re.finditer(pattern, text)
            for match in matches:
                if '도시근로자' in pattern:
                    income_assets["소득_기준"].append(f"도시근로자 {match.group(1)}%")
                elif '연소득' in pattern:
                    amount = self._convert_korean_amount(match.group(1))
                    income_assets["소득_기준"].append(f"연소득 {amount}만원")
                else:
                    household_type = match.group(1) if match.group(1) else ""
                    percentage = match.group(2)
                    income_assets["소득_기준"].append(f"중위소득 {percentage}% ({household_type})")
            
        # 재산 기준 추출 (기존 코드 유지)
        match = re.search(r'재산\s*(\d+)(?:백만|천만|억)원', text)
        if match:
            income_assets["재산_기준"] = self._convert_korean_amount(match.group(1))
            
        return income_assets

    def _convert_korean_amount(self, amount_str: str) -> int:
        # 숫자 변환 로직 구현
        try:
            return int(amount_str)
        except ValueError:
            return 0

    def extract_status(self, text: str) -> Dict[str, bool]:
        status = {
            "대학생": False,
            "직장인": False,
            "창업자": False,
            "귀농인": False
        }
        
        if '대학생' in text or '재학생' in text:
            status["대학생"] = True
        if '직장' in text or '근로자' in text:
            status["직장인"] = True
        if '창업' in text:
            status["창업자"] = True
        if '귀농' in text:
            status["귀농인"] = True
            
        return status

    def extract_requirements(self, text: str) -> Dict[str, bool]:
        requirements = {
            "청약통장": False,
            "전입신고": False,
            "부모별도거주": False
        }
        
        if '청약통장' in text:
            requirements["청약통장"] = True
        if '전입신고' in text:
            requirements["전입신고"] = True
        if '부모' in text and ('별도' in text or '따로' in text):
            requirements["부모별도거주"] = True
            
        return requirements

    
    def parse(self, item: Dict) -> PolicyCriteria:
        # 기본 정보 추출 (이전 코드의 기능 활용)
        text = self._get_detail(item, "거주지 및 소득")
        age_min, age_max = self.extract_age(text)
        regions = self.extract_regions(text)
        marriage_required = self.extract_marriage_status(text)
        residence_status = self.extract_residence_status(text)
        income_assets = self.extract_income_assets(text)
        status = self.extract_status(text)
        requirementss = self.extract_requirements(text)
        
        # 기타 조건은 향후 확장성을 위해 리스트로 유지
        # other = []
        
        # 새로운 정보 추출
        income_requirements = self.income_parser.parse_income_requirements(text)
        application_period = self.date_parser.parse_period(text)
        
        # 지원금액 추출
        support_amount = None
        if "지원금액" in text or "지원한도" in text:
            amount_text = text[text.find("지원금액"):]
            support_amount = self.monetary_parser.parse_amount(amount_text)
        
        # 요구사항 생성
        requirements = PolicyCriteria(
            age_min=age_min,
            age_max=age_max,
            region=regions,
            marriage_required=marriage_required,
            residence_status=residence_status,  # 이전 코드의 기능 활용
            income_requirements=income_requirements,
            asset_limit=None,  # 자산 제한 추출 필요
            application_period=application_period,
            support_amount=support_amount,
            requirements=requirementss,  # 이전 코드의 기능 활용
            validation_errors=[],
            income_assets=income_assets,
            status=status,
        )
        
        if debug:
            print(text)
            print(requirements)
            print('-'*50)

        # 검증 수행
        requirements.validation_errors = self.validator.validate_requirements(requirements)
        
        return requirements


@dataclass
class Policy:
    policy_id: str
    title: str
    criteria: PolicyCriteria
    description: str
    details: Dict[str, str]

class PolicyFilter:
    def __init__(self, json_data: str):
        self.policies = self._parse_policies(json_data)
    
    def _parse_policies(self, json_data: str) -> List[Policy]:
        data = json.loads(json_data)
        policies = []
        parser = ExtendedRequirementsParser()
        
        for item in data:
            criteria=parser.parse(item)
            
            policy = Policy(
                policy_id=self._get_detail(item, "정책 번호"),
                title=item["Policy Title"],
                criteria=criteria,
                description=item["Description"],
                details={detail["Title"]: detail["Content"] for detail in item["Details"]}
            )
            policies.append(policy)
            
        
        return policies
    
    def _get_detail(self, item: Dict, title: str) -> str:
        for detail in item["Details"]:
            if detail["Title"] == title:
                return detail["Content"]
        return ""
    
    
    def filter_policies(self, 
                       age: int = None,
                       region: str = None,
                       income_percent: float = None,
                       monthly_rent: int = None,
                       deposit: int = None,
                       is_married: bool = None) -> List[Policy]:
        """
        주어진 조건에 맞는 정책들을 필터링하여 반환합니다.
        
        Args:
            age: 나이
            region: 지역
            income_percent: 중위소득 대비 비율 (%)
            monthly_rent: 월세 금액 (만원)
            deposit: 보증금 금액 (만원)
            is_married: 혼인 여부
        """
        eligible_policies = []
        
        for policy in self.policies:
            criteria = policy.criteria
            
            # 각 조건 검사
            if age is not None and criteria.age_min is not None and criteria.age_max is not None:
                if not (criteria.age_min <= age <= criteria.age_max):
                    continue
                    
            # if region is not None and criteria.region is not None:
            #     if criteria.region != region:
            #         continue
                    
            # if income_percent is not None and criteria.income is not None:
            #     if income_percent > criteria.income_limit:
            #         continue
                    
            # if monthly_rent is not None and criteria.monthly_rent_limit is not None:
            #     if monthly_rent > criteria.monthly_rent_limit:
            #         continue
                    
            # if deposit is not None and criteria.deposit_limit is not None:
            #     if deposit > criteria.deposit_limit:
            #         continue
                    
            # if is_married is not None and criteria.marriage_required is not None:
            #     if criteria.marriage_required and not is_married:
            #         continue
            
            eligible_policies.append(policy)
            
        return eligible_policies

def main():
    # 사용 예시
    with open('filtered_policies.json', 'r', encoding='utf-8') as f:
        json_data = f.read()
    
    policy_filter = PolicyFilter(json_data)
    
    # 예시: 28세, 경북 거주, 중위소득 100%, 월세 50만원, 보증금 3000만원, 기혼
    eligible_policies = policy_filter.filter_policies(
        age=28,
        region="경북",
        income_percent=100,
        monthly_rent=50,
        deposit=3000,
        is_married=True
    )

    print(len(eligible_policies))
    
    for policy in eligible_policies:
        print(f"\n정책명: {policy.title}")
        print(f"정책 ID: {policy.policy_id}")
        print(f"지원 내용: {policy.details.get('거주지 및 소득', '')}")

if __name__ == "__main__":
    main()
```

```python
# 사업 운영 기간 파서
import re

# 날짜 패턴별 정규표현식
date_patterns = {
    # 1. YYYY.MM.DD. ~ YYYY.MM.DD. 형식
    'full_date_dots': r'(\d{4})\.(\d{1,2})\.(\d{1,2})\.?\s*~\s*(\d{4})\.(\d{1,2})\.(\d{1,2})\.?',
    
    # 2. YYYY. MM. ~ YYYY. MM. 형식
    'year_month_dots': r'(\d{4})\.\s*(\d{1,2})\.\s*~\s*(\d{4})\.\s*(\d{1,2})\.',
    
    # 3. YYYY-MM-DD~YYYY-MM-DD 형식
    'full_date_hyphens': r'(\d{4})-(\d{2})-(\d{2})\s*~\s*(\d{4})-(\d{2})-(\d{2})',
    
    # 4. YYYY. M. ~ MM. 형식 (같은 해 다른 달)
    'same_year_months': r'(\d{4})\.\s*(\d{1,2})\.\s*~\s*(\d{1,2})\.',
    
    # 5. 'YY. ~ 'YY. 형식
    'short_years': r'\'(\d{2})\.\s*~\s*\'(\d{2})\.',
    
    # 6. 'YY. MM. ~ 'YY. MM. 형식
    'short_year_month': r'\'(\d{2})\.\s*(\d{1,2})\.\s*~\s*\'(\d{2})\.\s*(\d{1,2})\.',
    
    # 7. YYYY년 MM월 ~ YYYY년 MM월 형식
    'korean_date': r'(\d{4})[년]\s*(\d{1,2})[월]\s*~\s*(\d{4})[년]\s*(\d{1,2})[월]',
    
    # 8. YYYY.M ~ YYYY.MM 형식
    'year_month_minimal': r'(\d{4})\.(\d{1,2})\s*~\s*(\d{4})\.(\d{1,2})',
    
    # 9. [NEW] YYYY. ~ YYYY. 형식 (연도만)
    'years_only': r'(\d{4})\.\s*~\s*(\d{4})\.',
    
    # 10. [NEW] YYYY. MM. DD.(요일) 형식
    'date_with_day': r'(\d{4})\.\s*(\d{1,2})\.\s*(\d{1,2})\.\s*\([월화수목금토일]\)',
    
    # 11. [NEW] 특수 케이스 패턴
    'special_cases': r'(연중|예산소진시까지|계속사업|현재)'
}

def parse_date_string(date_str):
    """
    주어진 날짜 문자열에서 시작일과 종료일을 추출
    
    Args:
        date_str (str): 날짜 문자열
    
    Returns:
        dict: 파싱된 날짜 정보 또는 특수 케이스 문자열
    """
    # 특수 케이스 먼저 체크
    special_match = re.search(date_patterns['special_cases'], date_str)
    if special_match:
        return {'type': 'special_case', 'value': special_match.group(1)}
    
    for pattern_name, pattern in date_patterns.items():
        if pattern_name == 'special_cases':
            continue
            
        match = re.search(pattern, date_str)
        if match:
            groups = match.groups()
            
            # 패턴별 결과 처리
            if pattern_name == 'full_date_dots':
                return {
                    'type': 'full_date',
                    'start': {'year': groups[0], 'month': groups[1], 'day': groups[2]},
                    'end': {'year': groups[3], 'month': groups[4], 'day': groups[5]}
                }
            elif pattern_name in ['year_month_dots', 'korean_date', 'year_month_minimal']:
                return {
                    'type': 'year_month',
                    'start': {'year': groups[0], 'month': groups[1]},
                    'end': {'year': groups[2], 'month': groups[3]}
                }
            elif pattern_name == 'full_date_hyphens':
                return {
                    'type': 'full_date',
                    'start': {'year': groups[0], 'month': groups[1], 'day': groups[2]},
                    'end': {'year': groups[3], 'month': groups[4], 'day': groups[5]}
                }
            elif pattern_name == 'same_year_months':
                return {
                    'type': 'same_year_months',
                    'start': {'year': groups[0], 'month': groups[1]},
                    'end': {'year': groups[0], 'month': groups[2]}
                }
            elif pattern_name == 'short_years':
                start_year = '20' + groups[0]
                end_year = '20' + groups[1]
                return {
                    'type': 'years_only',
                    'start': {'year': start_year},
                    'end': {'year': end_year}
                }
            elif pattern_name == 'short_year_month':
                start_year = '20' + groups[0]
                end_year = '20' + groups[2]
                return {
                    'type': 'year_month',
                    'start': {'year': start_year, 'month': groups[1]},
                    'end': {'year': end_year, 'month': groups[3]}
                }
            elif pattern_name == 'years_only':
                return {
                    'type': 'years_only',
                    'start': {'year': groups[0]},
                    'end': {'year': groups[1]}
                }
            elif pattern_name == 'date_with_day':
                return {
                    'type': 'single_date_with_day',
                    'date': {'year': groups[0], 'month': groups[1], 'day': groups[2]}
                }
    
    return None

# 테스트
test_dates = [
    "2024.11.05.~2024.12.27.",
    "2024. 2. ~ 2025. 2.",
    "2024-01-01~2024-12-31",
    "2024.1. ~ 12.",
    "'19. 4. ~ '26. 8.",
    "2024년 1월 ~ 2024년 12월",
    "2024.1 ~ 2024.12",
    "2017. ~ 2026.",
    "연중",
    "예산소진시까지",
    "2024. 2. 25.(화)"
]

for date in test_dates:
    result = parse_date_string(date)
    if result:
        print(f"\n입력: {date}")
        print(f"결과: {result}")
```

```python
import re

def parse_operating_periods(text):
    # Regular expression to match date ranges
    date_range_pattern = r"(\d{4})년\s*(\d{1,2})월\s*(\d{1,2})일\s*~\s*(\d{4})년\s*(\d{1,2})월\s*(\d{1,2})일"
    # Non-date cases like '상시' and '미정'
    status_pattern = r"상시|미정"

    results = []

    for line in text.splitlines():
        # Check for date ranges
        date_match = re.search(date_range_pattern, line)
        if date_match:
            start_year, start_month, start_day, end_year, end_month, end_day = date_match.groups()
            results.append({
                "start_date": f"{start_year}-{int(start_month):02d}-{int(start_day):02d}",
                "end_date": f"{end_year}-{int(end_month):02d}-{int(end_day):02d}"
            })
        else:
            # Check for '상시' or '미정'
            status_match = re.search(status_pattern, line)
            if status_match:
                results.append({
                    "start_date": "0000-0-0",
                    "end_date": "9999-99-99"
                })

    return results

# Example input text
text = """상시
2024년 11월  18일 ~ 2024년 12월  31일
미정
2024년 02월  26일 ~ 2025년 02월  25일
"""

# Parse and print the results
parsed_data = parse_operating_periods(text)
for entry in parsed_data:
    print(entry)

```

```python
# 나이 추출
import re

def extract_age_range(text):
    # 정규 표현식 정의
    age_pattern = r"(?:(?:만\s*)?(\d+)\s*세\s*~\s*(\d+)\s*세|(?:만\s*)?(\d+)\s*세\s*~\s*제한 없음|(?:만\s*)?제한 없음)"
    match = re.search(age_pattern, text)

    if match:
        if match.group(1) and match.group(2):  # "만 19세 ~ 34세" 형태
            min_age = int(match.group(1))
            max_age = int(match.group(2))
        elif match.group(3):  # "만 19세 ~ 제한 없음" 형태
            min_age = int(match.group(3))
            max_age = 9999
        else:  # "제한 없음" 형태
            min_age = 0
            max_age = 9999
    else:
        # 매치되지 않을 경우 기본 값
        min_age = 0
        max_age = 0

    return min_age, max_age

# 테스트
examples = [
    "만 19세 ~ 34세",
    "만 25세 ~ 제한 없음",
    "제한 없음",
    "만 18세 ~ 만 30세"
]

for example in examples:
    print(f"{example}: {extract_age_range(example)}")
```
